```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("/Users/koushik/Desktop/hackathon/house-prices-advanced-regression-techniques/train.csv")
print(df.shape)
print(df.head())
```

```python
#check for missing value percentage
#list of features with not a number value
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]
#we are trying to find the percentage of missing values that
for features in features_with_na:
    print(features,np.round(df[features].isnull().mean(),4),"%")#this line will print the percentage of missing value rounded off to 4 decimal places
```

```python
#we try to find the relation between sales price and missing values
for feature in features_with_na:
    data=df.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)#making a variable that indicates 1 for missing data and 0 for the rest
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
#list of numerical variables
numerical_features=[feature for feature in df.columns if df[feature].dtype!="O"]
print("Number of numerical variables : ",len(numerical_features))
df[numerical_features].head()
```

```python
#datetime variables(features that contain year information)
year_feature=[feature for feature in numerical_features if "Yr" in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    if feature !='YrSold':
        data=df.copy()
        #we are calculating the difference between the year variable and year the house was sold for as this would make a lot more sense than
        data[feature]=data["YrSold"]-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.show()
```

```python
#these graphs provides more info regarding year parameter

#checking for discreate variables
discrete_variables =[feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_feature+['Id']]#setting a random threshold value as 25 for number of unique values
print("Discrete variable count : ",len(discrete_variables))
```

```python
#finding the relationship between discrete features and sales price
for feature in discrete_variables:
    data=df.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
#from these graphs we can also say that there is relation between sale price and discrete variables
```

```python
#continuous variables
continuous_variables =[feature for feature in numerical_features if feature not in discrete_variables+year_feature+['Id']]
print("Continuous variable count : ",len(continuous_variables))
```

```python
#Analysing continuous variables
for feature in continuous_variables:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
#some datas are skewed
```

```python
#performing logarithmic transformation
for feature in continuous_variables:
    data=df.copy()
    if 0 in data[feature].unique():#log 0 is not defined
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data["SalePrice"])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel("feature")
        plt.ylabel("sales price")
        plt.title(feature)
        plt.show()
#we can see a positive corelation between the features
```

```python
#outliers
for feature in continuous_variables:
    data=df.copy()
    if 0 in data[feature].unique():#log 0 is not defined
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
#we can see that there are quite a lot of outliers present
```

```python
#missing values
#we are checking for categorical features which are missing 
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
for feature in features_nan:
    print(f"{feature} : {np.round(df[feature].isnull().mean(),4)} % missing values")#rounding off to 4 decimal places
```

```python
#Replace missing value with a new label
def replace_cat_feature(dataset,features):
    data=dataset.copy()
    data[features]=data[features].fillna("Missing")
    return data
df=replace_cat_feature(df,features_nan)
df[features_nan].isnull().sum()
#we can see that there are no nan value now
```

```python
#checking for numeric values now
numeric_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']
for feature in numeric_nan:
    print(f"{feature} : {np.round(df[feature].isnull().mean(),4)} % missing values")#rounding off to 4 decimal places
```

```python
#replacing numerical missing values
for feature in numeric_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numeric_nan].isnull().sum()
```

```python
#Temporal variables(Date time variables)
for feature in ["YearBuilt","YearRemodAdd","GarageYrBlt"]:
    df[feature]=df['YrSold']-df[feature]
df.head()
```

```python

num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
#We will remove categorical variables that are present less than 1% of the observations
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
categorical_features
```

```python
for feature in categorical_features:#325
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
dataset=pd.read_csv("x.csv")
dataset.head()
```

```python
#capture the dependent feture
y=dataset[["SalePrice"]]
```

```python
#drop dependent feature
x=dataset.drop(['Id',"SalePrice"],axis=1)
#x_test.head()
```

```python
#apply feature selection
#applying lasso regression
feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(x,y)
```

```python
feature_sel_model.get_support()
```

```python
#printing the number of total and selected features
selected_feat = x.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((x.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
selected_feat
```

```python
x=x[selected_feat]
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x, y, test_size=0.15)
```

```python
#x=x_train.loc[:]
#y=y_train.loc[:]
print(len(xtrain),len(xtest))
```

```python
xgbr=XGBRegressor(verbosity=0)
xgbr.fit(xtrain, ytrain)
```

```python
score = xgbr.score(xtrain, ytrain)  
print("Training score: ", score)
```

```python
ypred = xgbr.predict(xtest)
mse = mean_squared_error(ytest, ypred)
print("MSE: %.2f" % mse)
```

```python
x_ax = range(len(ytest))
plt.plot(x_ax, ytest, label="original")
plt.plot(x_ax, ypred, label="predicted")
plt.title("California test and predicted data")
plt.legend()
plt.show()
```
